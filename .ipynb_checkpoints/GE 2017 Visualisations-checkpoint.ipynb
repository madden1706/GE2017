```python
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
plt.style.use('fivethirtyeight')
#import seaborn; seaborn.set()

%matplotlib inline  
```

```python
GE2017 = pd.read_csv("GE2017.csv", header=0)
GE2017 = GE2017[GE2017.country_name != "Northern Ireland"]
GE2017 = GE2017[GE2017.first_party != "Spk"]
                     
print(GE2017.columns)
print(GE2017.shape)
```

I dropped the Northern Ireland seats and Speaker's seat as there is no Labour opposition in these seats. 

```python
#print(GE2017.loc[0])
```

```python
Lab_win = GE2017[GE2017["first_party"] == "Lab"]
#print(Lab_win)
```

```python
print(pd.unique(GE2017["first_party"]))
```

```python
Con_win = GE2017[GE2017["first_party"] == "Con"]
SNP_win = GE2017[GE2017["first_party"] == "SNP"]
LD_win = GE2017[GE2017["first_party"] == "LD"]
PC_win = GE2017[GE2017["first_party"] == "PC"]
Green_win = GE2017[GE2017["first_party"] == "Green"]
```

So I made a dataframes containing each of the seats won by a party and then plotted the majority over the Labour vote for each.

```python
plt.bar(Lab_win.index, Lab_win["majority"], color="r", align = "center", label="Lab")
plt.bar(Con_win.index, Con_win["lab"]-Con_win["con"], color = "b", align = "center", label="Con")
plt.bar(SNP_win.index, SNP_win["lab"]-SNP_win["snp"], color = "orange", align = "center", label="SNP")
plt.bar(LD_win.index, LD_win["lab"]-LD_win["ld"], color = "yellow", align = "center", label="LD")
plt.bar(PC_win.index, PC_win["lab"]-PC_win["pc"], color = "green", align = "center", label="PC")
plt.bar(Green_win.index, Green_win["lab"]-Green_win["green"], color = "lawngreen", align = "center", label="Green")

ax = plt.subplot()
box = ax.get_position()
#plt.legend()
ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
# Put a legend to the right of the current axis
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))


plt.title("Labour Majority - GE2017")
plt.ylabel("Labour Majority")
plt.xlabel("Constituency A-Z")
plt.tick_params(axis="x",  labelbottom='off', bottom="off")

#plt.savefig("LabourMajorityGE2017.png", dpi=300, pad_inches=1)
```

The above plot was a little messy, so I sorted by majority to give a cleaer picture. 

```python
GE2017["Lab_majority"] = 0

GE2017["Lab_majority"] = np.where(GE2017["first_party"] == "Lab", GE2017["majority"], 0)


GE2017["Con_majority"] = Con_win["lab"]-Con_win["con"]
GE2017["SNP_majority"] =  SNP_win["lab"]-SNP_win["snp"]
GE2017["LD_majority"] =  LD_win["lab"]-LD_win["ld"]
GE2017["PC_majority"] = PC_win["lab"]-PC_win["pc"]
GE2017["Green_majority"] = Green_win["lab"]-Green_win["green"]
#print(GE2017.head())

#df['e'] = df[col_list].sum(axis=1)

GE2017["Sum_Lab_majority"] = GE2017[["Lab_majority","Con_majority", "SNP_majority", "LD_majority", "PC_majority", "Green_majority"]].sum(1)

#print(GE2017.head())
```

```python
GE2017_sort = GE2017.sort_values(by=["Sum_Lab_majority"], ascending = False)
GE2017_sort = GE2017_sort.reset_index()
#print(GE2017_sort.head())

colours_df = GE2017_sort[["constituency_name", "first_party"]]
#print(colours_df)
colours_df.set_index("constituency_name", inplace=True)
colours_dict = colours_df.to_dict("series")
#print(colours_dict)
party_colours = {"Lab" : "red", "Con" : "blue", "SNP": "yellow", "LD" : "orange", "PC" : "green", "Green": "lawngreen"}
GE2017_sort["Colour"] = GE2017_sort.loc[:,"first_party"].map(party_colours)
```

```python
plt.bar(GE2017_sort.index, GE2017_sort["Sum_Lab_majority"], align = "center", color =GE2017_sort.Colour)
#plt.bar(SNP_win.index, SNP_win["lab"]-SNP_win["snp"], color = "orange", align = "center")
#plt.bar(LD_win.index, LD_win["lab"]-LD_win["ld"], color = "yellow", align = "center")
#plt.bar(PC_win.index, PC_win["lab"]-PC_win["pc"], color = "green", align = "center")

plt.title("Labour Majority - GE2017")
plt.ylabel("Labour Majority")
plt.xlabel("Constituency")
plt.tick_params(axis="x",  labelbottom='off', bottom="off")

Lab = mpatches.Patch(color='red', label='Lab')
Con = mpatches.Patch(color='blue', label='Con')
SNP = mpatches.Patch(color='yellow', label='SNP')
LD = mpatches.Patch(color='orange', label='LD')
PC = mpatches.Patch(color='green', label='PC')
Green = mpatches.Patch(color='lawngreen', label='Green')
plt.legend(handles=[Lab, Con, SNP, LD, PC, Green], )


plt.savefig("LabourMajorityGE2017_1.png", dpi=900, pad_inches=1)
```

The above graph is a little too busy, so a focused graph of the 35 seats with the smallest positive and negative Labour majorities is below.

```python

GE_temp = GE2017_sort["Lab_majority"] > 0
X = (GE_temp == 1).sum()
GE2017_sort_slice = GE2017_sort[int(X-35) : int(X+35) ]
GE2017_sort_slice["Colour"] = GE2017_sort_slice.loc[:,"first_party"].map(party_colours)


plt.bar(GE2017_sort_slice.index, GE2017_sort_slice["Sum_Lab_majority"], align = "center", color =GE2017_sort_slice.Colour,
       tick_label=GE2017_sort_slice["constituency_name"])
plt.title("Labour Majority - GE2017")
plt.ylabel("Labour Majority")
plt.xlabel("Constituency")

Lab = mpatches.Patch(color='red', label='Lab')
Con = mpatches.Patch(color='blue', label='Con')
SNP = mpatches.Patch(color='yellow', label='SNP')
LD = mpatches.Patch(color='orange', label='LD')
PC = mpatches.Patch(color='green', label='PC')
Green = mpatches.Patch(color='lawngreen', label='Green')
plt.legend(handles=[Lab, Con, SNP, LD, PC, Green], )


plt.tick_params(axis="x", rotation =90, labelsize=3, width = 1)
#plt.tick_label(GE2017_sort_slice["constituency_name"].tolist(), rotation='vertical')
#plt.tick_params(axis='x', labelsize=5, rotation  = 75 )
plt.tight_layout()

plt.savefig("LabourMajorityGE2017_1.2.png", dpi=300, pad_inches=1)



```
